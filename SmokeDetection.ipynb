```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# Load the trained model
model_path = "/content/best.pt"  # Update if needed
model = YOLO(model_path)

# Path to your image
image_path = "/content/FishLake-burn2017__KellyCornwallUSFS.jpg"  # Update with your image path

# Perform prediction
results = model(image_path)

# Get the image and predictions
img = results[0].orig_img  # Get original image
detections = results[0].boxes.data  # Get detection data

# Draw bounding boxes and labels on the image
for detection in detections:
    x1, y1, x2, y2, conf, cls = detection
    x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
    label = f"{model.names[int(cls)]} {conf:.2f}"
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Draw rectangle
    cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)  # Add label

# Display the image
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
import numpy as np

# Load the trained model
model_path = "/content/best.pt"  # Update with your model path
model = YOLO(model_path)

# Path to your video file
video_path = "/content/204361-924209264_medium.mp4"  # Update with your video path

# Open the video file
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get video properties for output
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Optional: Define a VideoWriter to save output
output_path = "/content/output_video.avi"
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'XVID'), fps, (frame_width, frame_height))

while cap.isOpened():
    # Read a frame from the video
    ret, frame = cap.read()

    if not ret:
        break  # Break the loop if the video has ended

    # Perform prediction on the frame
    results = model(frame)

    # Draw bounding boxes and labels on the frame
    for detection in results[0].boxes.data:
        x1, y1, x2, y2, conf, cls = detection.cpu().numpy()
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
        label = f"{model.names[int(cls)]} {conf:.2f}"
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Draw rectangle
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)  # Add label

    # Display the frame (or write it to a file)
    try:
        from google.colab.patches import cv2_imshow
        cv2_imshow(frame)  # Use cv2_imshow for Colab
    except ImportError:
        cv2.imshow("YOLO Detection", frame)  # Use cv2.imshow for local execution

    # Save the frame to output video
    out.write(frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and writer
cap.release()
out.release()
cv2.destroyAllWindows()
```
